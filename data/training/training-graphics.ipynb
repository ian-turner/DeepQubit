```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
# loading data
names = ['Matrix', 'Matrix + NeRF', 'Quaternion + NeRF', 'Hurwitz + NeRF']
data = [np.loadtxt(x) for x in [
    '1q1e-2-P.txt',
    '1q1e-2-P-L15.txt',
    '1q1e-2-Q-P-L15.txt', 
    '1q1e-2-H-P-L15.txt',
]]

# plot setup
fig, ax = plt.subplots()
ax.set_title(r'Training progress (1 qubit, $\epsilon=0.01$)')
ax.set_xlabel('Training iteration')
ax.set_ylabel('Percent of test states solved')

for (name, ys) in zip(names, data):
    ax.plot(ys, label=name)

# plot customization
ax.legend()

#plt.savefig('training1q.png')
plt.show()
```
