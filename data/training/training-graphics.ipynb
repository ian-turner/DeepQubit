```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading data
names = ['Base', 'Quaternion', 'Hurwitz']
data = [np.loadtxt(x) for x in [
    '1q1e-2-base.txt', 
    '1q1e-2-Q-P-L15.txt', 
    '1q1e-2-H-P-L15.txt',
]]

# plot setup
fig, ax = plt.subplots()
ax.set_title('Training progress')
ax.set_xlabel('Training iteration')
ax.set_ylabel('Percent of test states solved')

for (name, ys) in zip(names, data):
    ax.plot(ys, label=name)

# plot customization
ax.legend()

plt.show()
```
