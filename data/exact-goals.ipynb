```python
import pickle
from domains.qcircuit import *
from utils.matrix_utils import *
```

```python
CNOT = CNOTGate(2, 0, 1).unitary
CZ = CZGate(2, 0, 1).unitary
CH = CHGate(2, 0, 1).unitary
```

```python
goal_matrices = [CNOT, CZ, CH]
goals = [QGoal(np.array(x, dtype=np.complex128)) for x in goal_matrices]
I2 = tensor_product([I, I])
states = [QState(I2) for _ in range(len(goal_matrices))]
```

```python
data = {'states': states, 'goals': goals}
pickle.dump(data, open('./n2_goals.pkl', 'wb'))
```
